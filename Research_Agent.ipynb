```python
import warnings
warnings.filterwarnings('ignore')
import os
import anthropic
from ddgs import DDGS

client = anthropic.Anthropic(
    api_key=userdata.get('ANTHROPIC_API_KEY')
)

def search_web(query):
    with DDGS() as ddgs:
        results = list(ddgs.text(query, max_results=3))
    text = ""
    for r in results:
        text += f"Title: {r['title']}\nSummary: {r['body']}\n\n"
    return text

def research_agent(topic):
    print(f"Searching for: {topic}\n")
    search_results = search_web(topic)
    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1024,
        system="You are an expert research assistant. Summarize clearly in plain text only.",
        messages=[{
            "role": "user",
            "content": f"Search results about '{topic}':\n\n{search_results}\n\nProvide a clear summary."
        }]
    )
    return response.content[0].text

print("Research Agent Ready!")
print(research_agent("Latest AI agent frameworks 2025"))
```

```python
# Test with different topics
print("Test 1:")
print(research_agent("Benefits of AI in education"))

print("\nTest 2:")
print(research_agent("Cryptography and machine learning"))

print("\nTest 3:")
print(research_agent("Houston Texas technology jobs 2025"))
```
